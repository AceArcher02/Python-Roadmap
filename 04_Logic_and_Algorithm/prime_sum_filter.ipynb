```python
# A list of mixed numbers
numbers = [10, 7, 4, 13, 21, 5, 8, 2, 11]
prime_numbers = []
total_sum = 0

print("Analyzing list:", numbers)

for num in numbers:
    # 1 is not prime, and we only care about positive numbers
    if num > 1:
        is_prime = True
        for i in range(2, int(num**0.5) + 1): # Optimized: only check up to square root
            if (num % i == 0):
                is_prime = False
                break

        if is_prime:
            prime_numbers.append(num)
            total_sum += num

print("Primes found in list:", prime_numbers)
print("The sum of these primes is:", total_sum)
```
