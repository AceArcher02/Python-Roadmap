```python
list1 = [1, 3, 4, 6, 8]
list2 = [4, 5, 6, 2, 10]
res = []

# Using len(list1) to find the number of items
for i in range(0, len(list1)):
    res.append(list1[i] + list2[i])

print(res) # result
```
